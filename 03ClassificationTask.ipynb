```python
import pandas as pd
import os

# Set the path to the extracted dataset folder
dataset_path = '/content/spam_2'

# Create empty lists to store the email content and labels
emails = []
labels = []

# Iterate through the extracted files
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        file_path = os.path.join(root, file)

        # Read the email content and label
        with open(file_path, 'r', encoding='latin1') as f:
            content = f.read()

        # Determine the label based on the folder name
        label = os.path.basename(root)

        # Append the email content and label to the lists
        emails.append(content)
        labels.append(label)

del emails[250:]
del labels[250:]

dataset_path = '/content/hard_ham'
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        file_path = os.path.join(root, file)

        # Read the email content and label
        with open(file_path, 'r', encoding='latin1') as f:
            content = f.read()

        # Determine the label based on the folder name
        label = os.path.basename(root)

        # Append the email content and label to the lists
        emails.append(content)
        labels.append(label)

# Create a pandas DataFrame from the lists
data = pd.DataFrame({'email': emails, 'label': labels})

# Shuffle the data (optional)
data = data.sample(frac=1, random_state=42).reset_index(drop=True)

# Print the first few rows of the DataFrame
print(data.head())
```

```python
import string
import re
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer

def clean(text):
  punc_free = text.translate(str.maketrans("", "", string.punctuation))
  lower = punc_free.lower()
  text = re.sub(r'\d+', '', lower)
  tokens = word_tokenize(text)
  stop_words = set(stopwords.words('english'))
  filtered_words = [word for word in tokens if word.lower() not in stop_words]
  stemmer = PorterStemmer()
  lemmatizer = WordNetLemmatizer()
  stemmed_words = [stemmer.stem(token) for token in filtered_words]
  lemmatized_words = [lemmatizer.lemmatize(token) for token in stemmed_words]
  filtered_text = ' '.join(lemmatized_words)
  return filtered_text
```

```python
X_train_clean = []
for text in X_train:
  X_train_clean.append(clean(text))
X_test_clean = []
for text in X_test:
  X_test_clean.append(clean(text))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
train_features = tfidf_vectorizer.fit_transform(X_train_clean)
test_features = tfidf_vectorizer.transform(X_test_clean)
```

```python
X_train_data = train_features.toarray()
X_test_data = test_features.toarray()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score

param_grid = {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6]}
grid_search = GridSearchCV(estimator = model, param_grid=param_grid, scoring='f1')
cv_scores = cross_val_score(grid_search, X_train_data, y_train, cv=5, scoring='f1')
grid_search.fit(X_train_data, y_train)
best_model = grid_search.best_estimator_
print(best_model)
y_pred_test = best_model.predict(X_test_data)
precision = precision_score(y_test, y_pred_test)
recall = recall_score(y_test, y_pred_test)
f1 = f1_score(y_test, y_pred_test)
print("Cross-validation Precision Scores:", cv_scores)
print("Best var_smoothing value:", best_model.var_smoothing)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1)
```
